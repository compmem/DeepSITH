```python
linew=3
with sn.plotting_context("notebook", font_scale=2):

    fig, axes = plt.subplots(2,2,sharey=True, figsize=(12,12))

    ax = axes[0,0]
    few_dat = dat[dat["Signal Length"]=="100"]
    sn.lineplot(data=few_dat, x="Training Step", y="avg_loss",
                hue='Model', ax=ax, legend=False, linewidth=4)
    ax.set_ylim(0, .19)
    ax.set_xlim(0, 200)
    ax.set_title('T = 100')
    ax.grid()
    ax.set_ylabel("Loss")
    ax.set_xlabel("")

    ax = axes[0,1]
    few_dat = dat[dat["Signal Length"]=="500"]
    sn.lineplot(data=few_dat, x="Training Step", y="avg_loss",
                hue='Model', ax=ax, legend=False, linewidth=4)
    ax.set_ylim(0, .19)
    ax.set_xlim(0, 200)
    ax.set_title('T = 500')
    ax.grid()
    ax.set_ylabel("Loss")
    ax.set_xlabel("")

    ax = axes[1,0]
    few_dat = dat[dat["Signal Length"]=="2000"]
    sn.lineplot(data=few_dat, x="Training Step", y="avg_loss",
                hue='Model', ax=ax, legend=False, linewidth=4)
    ax.set_ylim(0, .19)
    ax.set_xlim(0, 200)
    ax.set_title('T = 2000')
    ax.grid()
    ax.set_ylabel("Loss")
    ax.set_xlabel("Training Step (Hundreds)")

    ax = axes[1,1]
    few_dat = dat[dat["Signal Length"]=="5000"]
    sn.lineplot(data=few_dat, x="Training Step", y="avg_loss",
                hue='Model', ax=ax, linewidth=4)
    ax.set_ylim(-.005, .19)
    ax.set_xlim(0, 200)
    ax.set_ylabel("Loss")
    ax.set_xlabel("Training Step (Hundreds)")

    ax.grid()
    ax.set_title('T = 5000')
    plt.savefig('AddingProblem.svg',
                #bbox='tight',
                edgecolor=fig.get_edgecolor(),
                facecolor=fig.get_facecolor(),
                dpi=150
                )
    plt.savefig('AddingProblem.pdf',
                #bbox='tight',
                edgecolor=fig.get_edgecolor(),
                facecolor=fig.get_facecolor(),
                dpi=150
                )
```

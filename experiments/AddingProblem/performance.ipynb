```python
from matplotlib import pylab as plt
import pandas as pd
import seaborn as sn
import os
sn.set_context("poster")
```

```python
# SITH
dat = pd.read_csv(os.path.join('perf', 'adding500_deepsith_8.csv'))
dat['Signal Length'] = ["500"]*dat.shape[0]
dat['Model'] = ["SITH"]*dat.shape[0]
dat['Training Step'] = (dat.batch_idx)/100

tdat = pd.read_csv(os.path.join('perf', 'adding2000_deepsith_3.csv'))
tdat['Signal Length'] = ["2000"]*tdat.shape[0]
tdat['Model'] = ["SITH"]*tdat.shape[0]
tdat['Training Step'] = (tdat.batch_idx)/100
dat = dat.append(tdat)
tdat = pd.read_csv(os.path.join('perf', 'adding5000_deepsith_3.csv'))
tdat['Signal Length'] = ["5000"]*tdat.shape[0]
tdat['Model'] = ["SITH"]*tdat.shape[0]
tdat['Training Step'] = (tdat.batch_idx)/100
dat = dat.append(tdat)

# LSTM
tdat = pd.read_csv(os.path.join('perf', 'adding500_lstm_3.csv'))
tdat['Signal Length'] = ["500"]*tdat.shape[0]
tdat['Model'] = ["LSTM"]*tdat.shape[0]
tdat['Training Step'] = (tdat.batch_idx)/100
dat = dat.append(tdat)
tdat = pd.read_csv(os.path.join('perf', 'adding2000_lstm_2.csv'))
tdat['Signal Length'] = ["2000"]*tdat.shape[0]
tdat['Model'] = ["LSTM"]*tdat.shape[0]
tdat['Training Step'] = (tdat.batch_idx)/100
dat = dat.append(tdat)
tdat = pd.read_csv(os.path.join('perf', 'adding5000_lstm_1.csv'))
tdat['Signal Length'] = ["5000"]*tdat.shape[0]
tdat['Model'] = ["LSTM"]*tdat.shape[0]
tdat['Training Step'] = (tdat.batch_idx)/100
dat = dat.append(tdat)

# LMU
tdat = pd.read_csv(os.path.join('perf', 'adding500_lmu_8.csv'))
tdat['Signal Length'] = ["500"]*tdat.shape[0]
tdat['Model'] = ["LMU"]*tdat.shape[0]
tdat['Training Step'] = (tdat.batch_idx)/100
dat = dat.append(tdat)
tdat = pd.read_csv(os.path.join('perf', 'adding2000_lmu_1.csv'))
tdat['Signal Length'] = ["2000"]*tdat.shape[0]
tdat['Model'] = ["LMU"]*tdat.shape[0]
tdat['Training Step'] = (tdat.batch_idx)/100
dat = dat.append(tdat)
tdat = pd.read_csv(os.path.join('perf', 'adding5000_lmu_1.csv'))
tdat['Signal Length'] = ["5000"]*tdat.shape[0]
tdat['Model'] = ["LMU"]*tdat.shape[0]
tdat['Training Step'] = (tdat.batch_idx)/100
dat = dat.append(tdat)
```

```python
fig = plt.figure(figsize=(20,10))

ax = fig.add_subplot(2,2,1)
few_dat = dat[dat["Signal Length"]=="500"]
sn.lineplot(data=few_dat, x="Training Step", y="avg_loss",
            hue='Model', ax=ax)
ax.set_ylim(0, .2)
ax.set_xlim(0, 200)
ax.set_ylabel("Loss")
ax.set_xlabel("")

ax = fig.add_subplot(2,2,2)
few_dat = dat[dat["Signal Length"]=="2000"]
sn.lineplot(data=few_dat, x="Training Step", y="avg_loss",
            hue='Model', ax=ax)
ax.set_ylim(0, .2)
ax.set_xlim(0, 200)
ax.set_ylabel("")
ax.set_xlabel("")

ax = fig.add_subplot(2,2,3)
few_dat = dat[dat["Signal Length"]=="5000"]
sn.lineplot(data=few_dat, x="Training Step", y="avg_loss",
            hue='Model', ax=ax)
ax.set_ylim(0, .2)
ax.set_xlim(0, 200)
ax.set_ylabel("Loss")
plt.savefig('AddingProblem.png',
            #bbox='tight',
            edgecolor=fig.get_edgecolor(),
            facecolor=fig.get_facecolor(),
            dpi=150
            )
```
